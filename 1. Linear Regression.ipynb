```python
#import labraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Import dataset
dataset = pd.read_csv('1. Salary_DataSet.csv')
```

```python
#Split column of Dependent and independent variable
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,1].values
```

```python
# Spliting data to training and testing dataset
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
#Regression
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train, Y_train)
```

```python
y_pridict = reg.predict(X_test)
```

```python
#now show the data in graph for testing data
plt.scatter(X_train, Y_train, color='red')
plt.plot(X_train, reg.predict(X_train), color='black')
plt.title("Linear Regression Experience VS Salary")
plt.show()
```

```python
#show graph for Testing data
plt.scatter(X_test, Y_test, color='red')
plt.plot(X_train, reg.predict(X_train), color='black')
plt.title("Linear Regression Experience VS Salary")
plt.show()
```
